```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow  # Use cv2.imshow if not in Colab

```

```python
# Load the image (grayscale)
image = cv2.imread('tiger.png', cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image, (400,400))

# Compute the histogram for the grayscale image
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
```

```python
# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.plot(hist)
plt.xlim([0, 256])  # Range of pixel intensity values
plt.show()
```

```python
# For color images (Red, Green, Blue channels)
color_image = cv2.imread('tiger.png')
color_image = cv2.resize(color_image, (400, 400))
colors = ('b', 'g', 'r')
plt.figure()
plt.title("Color Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

```

```python
# Define color channels
colors = ('b', 'g', 'r')

# Loop through each color channel
for i, color in enumerate(colors):
    hist = cv2.calcHist([color_image], [i], None, [256], [0, 256])  # Calculate histogram for each channel
    plt.plot(hist, color=color)  # Plot histogram for that channel
    plt.xlim([0, 256])  # Limit x-axis to intensity range [0, 256]

plt.show()  # Show the plot
```

```python
# Apply Histogram Equalization
equalized_image = cv2.equalizeHist(image)


# Plot histograms for comparison
plt.figure()
plt.title("Original vs Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Original image histogram
original_hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.plot(original_hist, label='Original', color='red')

# Equalized image histogram
equalized_hist = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
plt.plot(equalized_hist, label='Equalized', color='black')

plt.legend()
plt.xlim([0, 256])
plt.show()
```
