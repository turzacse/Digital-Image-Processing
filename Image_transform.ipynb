```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('tiger.png',cv2.IMREAD_GRAYSCALE)
image = cv2.resize(image,(400, 400))

negated_image = 255 - image

cv2_imshow(negated_image)
```

```python
# Logarithmic transformation
c = 255 / np.log(1 + np.max(image))  # Calculate scaling constant

log_transformed_image = c * (np.log(1 + image))
# log_transformed_image = np.array(log_transformed_image, dtype=np.uint8)

cv2_imshow(log_transformed_image)
```

```python
# Power-law transformation (Gamma correction)
gamma = 2.2  # Choose gamma value

# Normalize the image to the range [0, 1]
normalized_image = image / 255.0
# Apply the gamma correction
gamma_corrected_image = np.power(normalized_image, gamma)

# Scale back to [0, 255]
gamma_corrected_image = np.uint8(gamma_corrected_image * 255)

# Display the gamma corrected image
cv2_imshow(gamma_corrected_image)
```
