```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('tiger.png')
image = cv2.resize(image, (400, 400))

# Apply average filtering (mean filtering)
kernel_size = (10, 10)
average_filtered_image = cv2.blur(image, kernel_size)

cv2_imshow(average_filtered_image);
```

```python
gaussian_filtered_image = cv2.GaussianBlur(image, (5, 5), 3)

cv2_imshow(gaussian_filtered_image)
```

```python
median_filtered_image = cv2.medianBlur(image, 5)

cv2_imshow(median_filtered_image)
```
