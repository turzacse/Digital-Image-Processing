```python
image1 = cv2.imread('couple.png')
image2 = cv2.imread('couple2.png')

image1 = cv2.resize(image1, (300, 300))
image2 = cv2.resize(image2, (300, 300))
print('Image Number 01')
cv2_imshow(image1)
print('Image Number 02')
cv2_imshow(image2)
added_image = cv2.add(image1, image2);
print('The Added Image of 01 & 02')
cv2_imshow(added_image)
cv2.imwrite('Added_img.png', added_image);
```

```python
subtracted_image = cv2.subtract(image1, image2)

# Display the subtracted image
cv2_imshow(subtracted_image)
cv2.imwrite('Subtracted_img.png', added_image);
```

```python
blended_image = cv2.addWeighted(image1, 0.5, image2, 0.5, 0)

# Display the blended image
cv2_imshow(blended_image)
cv2.imwrite('blended_img.png', added_image);
```

```python
bitwise_and = cv2.bitwise_and(image1, image2)
cv2_imshow(bitwise_and)  # Display Bitwise AND result

cv2.imwrite('bitwise_and.png', bitwise_and);
```

```python
bitwise_or = cv2.bitwise_or(image1, image2)
cv2_imshow(bitwise_or)  # Display Bitwise or result

cv2.imwrite('bitwise_or.png', bitwise_or);
```

```python
bitwise_xor = cv2.bitwise_xor(image1, image2)
cv2_imshow(bitwise_xor)  # Display Bitwise xor result

cv2.imwrite('bitwise_xor.png', bitwise_xor);
```

```python
bitwise_not = cv2.bitwise_not(image1)
cv2_imshow(bitwise_not)  # Display Bitwise not result

cv2.imwrite('bitwise_not.png', bitwise_not);
```

```python
destination = np.zeros_like(image1)

mask = np.zeros(image1.shape[:2], dtype="uint8")
cv2.rectangle(mask, (75, 75), (225, 225), 255, -1)

bitwise_and = cv2.bitwise_and(image1, image2, destination, mask)
cv2_imshow(bitwise_and)

cv2.imwrite('bitwise_and_with_mask.jpg', bitwise_and)
```
