```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('couple.png', cv2.IMREAD_GRAYSCALE)

# cv2_imshow(image)

_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# cv2_imshow(binary_image)
```

```python
rows, cols = image.shape

newImg = np.zeros((rows, cols))
for i in range(rows):
  for j in range(cols):
    if image[i][j] > 127:
      newImg[i][j] = 255

cv2_imshow(newImg)
cv2.imwrite('Black_White.png', newImg)
```
