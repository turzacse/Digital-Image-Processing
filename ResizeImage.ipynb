```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('tiger.png')
fx = 0.4
fy = 0.4
interpolation = cv2.INTER_LINEAR

resized_image = cv2.resize(image, dsize=None, fx=fx, fy=fy, interpolation=interpolation)

cv2_imshow(resized_image)
```
